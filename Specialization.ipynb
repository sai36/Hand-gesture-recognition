```python
# Pre-processing the images
X = [] # Image data
y = [] # Labels
for filename in os.listdir(r'C:\\Users\\saiteja\\Desktop\\Masters\\Intelligent_Interactive_Systems\\project_req\\videos\\imgs'):
    if filename.endswith(".png") or filename.endswith(".py"):
        path = "imgs\\" + filename
        img = cv2.imread(path)# Reads image and returns np.array
        img= cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) # Converts into the corret colorspace (GRAY)
        img = cv2.resize(img, (320, 120)) # Reduce image size so training can be faster
        X.append(img)
        
        name = filename.split('-')
        # Processing label in image path
        #if (len(filename.split("_")) > 3):
            #category = filename.split("_")
        #else:
            #category = filename.split("_")[:2]
        #new = category.join()
        name = name[0]
        print (name)
        val = {'fist_dorsal' : 0,'fist_palm': 1,'open_dorsal' : 2,'open_palm': 3,'three_fingers_dorsal' : 4,'three_fingers_palm' : 5} 
        label = val[name]
        
        #if(category=='palm'):
            #label = [0]# We need to convert 10_down to 00_down, or else it crashes
        #else:
            #label = [1]
        y.append(label)
```

```python

X = np.array(X, dtype="uint8")
X = X.reshape(11997, 120, 320, 1) # Needed to reshape so CNN knows it's different images
y = np.array(y)
print("Images loaded: ", len(X))
print("Labels loaded: ", len(y))
```

```python
# Percentage of images that we want to use for testing. 
# The rest is used for training.

ts = 0.3
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=ts, random_state=42)
```

```python
# Architecture of the model
model = Sequential()
model.add(Conv2D(32, (5, 5), activation='relu', input_shape=(120, 320, 1))) 
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(64, (3, 3), activation='relu')) 
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dense(10, activation='softmax'))
# Configures the model for training
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
# Trains the model for a given number of epochs (iterations on a dataset) and validates it.
model.fit(X_train, y_train, epochs=5, batch_size=64, verbose=2, validation_data=(X_test, y_test))
```

```python
# Saving the model
model.save('subsystem_2.h5')
```

```python
# Accuracy
test_loss, test_acc = model.evaluate(X_test, y_test)

print('Test accuracy: {:2.2f}%'.format(test_acc*100))
```
